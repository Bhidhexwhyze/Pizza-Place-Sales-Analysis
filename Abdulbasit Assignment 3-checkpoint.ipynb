## Understanding the Relationship

| Table         | Key           | Joins With    |
| ------------- | ------------- | ------------- |
| orders        | order_id      | orders_details |
| orders_details | pizza_id      | pizzas        |
| pizzas        | pizza_type_id | pizza_types   |


```python
# Merging the datasets

# Merge orders + orders_details

df = pd.merge(orders_details, orders, on ='order_id')
```

```python
# Merge with pizzas

df = pd.merge(df, pizzas, on ='pizza_id')
```

```python
# Merge with pizza_types

df = pd.merge(df, pizza_types, on='pizza_type_id')
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
#Creating analysis column

df['day'] = df['date'].dt.day_name()
df['month'] = df['date'].dt.month_name()
df['hour'] = pd.to_datetime(df['time']).dt.hour
```

```python
#Creating revenue column

df["revenue"] = df["quantity"] * df["price"]

df["revenue"]
```

```python
#Peak sales hour

df.groupby("hour")["revenue"].sum().plot(kind="bar")
plt.title("Peak Sales Hours")
plt.show()
```

This idenify busiest hours

```python
#Sales by day (Identifying the highest day)

df.groupby("day")["revenue"].sum().plot(kind='bar')
plt.show()
```

```python
#Top 5 best selling pizza

df.groupby("name")["quantity"].sum().sort_values(ascending=False).head(5).plot(kind="bar")
plt.show()
```

```python
#Monthly sales tread

df.groupby("month")["revenue"].sum().plot(kind="line", marker = "o")
plt.show()
```

```python
#Worst performing pizza

df.groupby("name")["quantity"].sum().sort_values().head(5)
```

```python
#Revenue by category

df.groupby("category")["revenue"].sum()
```

```python
#Revenue by size

df.groupby('size')['revenue'].sum()
```

```python
#Average order value

df.groupby('order_id')['revenue'].sum().mean()
```
